```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
```

```python
image = io.imread('/content/main-qimg-dd1332ccf97d327693838984e86a51da-lq.jpg')
io.imshow(image)
io.show()
```

```python
init_c=image.shape[0]*image.shape[1]*image.shape[2]*8
print("Intial cost is: ",init_c," bits");
```

```python
rows = image.shape[0]
cols = image.shape[1]
```

```python
image = image.reshape(rows*cols, 3)
```

```python
k=64
```

```python
kmeans = KMeans(n_clusters=k)
kmeans.fit(image)
```

```python
compressed_image = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = np.clip(compressed_image.astype('uint8'), 0, 255)
```

```python
compressed_image = compressed_image.reshape(rows, cols, 3)
```

```python
io.imsave('compressed_image.png', compressed_image)
io.imshow(compressed_image)
io.show()
```

```python
final_c=compressed_image.shape[0]*compressed_image.shape[0]*np.log2(k)+k*3*8
print("Final Cost is: ",final_c," bits")
comp=1-final_c/init_c
print("Compression Factor is: ",100*comp)
```
